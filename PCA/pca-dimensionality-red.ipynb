```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data.csv')
df.head()
```

```python
df.columns
```

```python
data = df.drop(columns=['diagnosis','Unnamed: 32'],axis=1)
```

```python
data.info()
```

```python
X = data
```

```python
mean = np.mean(X, axis=0)
X_centered = X - mean
```

# centered data

```python
X_centered
```

# covarience b/w all 32 features 

```python
covariance_matrix = np.cov(X_centered, rowvar=False)
covariance_matrix.shape
```

# eigen values & vectors

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
print(f'igenvalues = {eigenvalues}, igenvectors = {eigenvectors} ')
```

```python
# Sort eigenvectors by decreasing eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

# Principle Component to retain

```python
num_components = 2
selected_eigenvectors = eigenvectors[:, :num_components]
```

```python
transformed_data = np.dot(X_centered, selected_eigenvectors)
```

# Transform the data

```python
transformed_data = np.dot(X_centered, selected_eigenvectors)

print("Original data:\n", X.shape)
print("Transformed data:\n", transformed_data.shape)
```

```python
from matplotlib import pyplot as plt
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.show()
```

# using SKLEARN

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the dataset
data = load_iris()
X = data.data
y = data.target

# Perform PCA
n_components = 2  # Choose the number of components
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.show()
```
